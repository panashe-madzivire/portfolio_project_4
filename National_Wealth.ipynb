```python
#to export the data frame to an sql data table
conn = pymysql.connect(host = 'localhost', database = 'covid_and_wealth', user = 'root', password = 'mADZIVIRE1996')
cursor = conn.cursor()
```

```python
def CreateTable(df,SQL_table):
    insert_query = f'INSERT INTO {SQL_table} VALUES'
    values_array = df.values
    df_values_export =[]

    for row in values_array:
        df_values_export.append(tuple(row))
    insert_query +=str(df_values_export)[1:-1] + ';'
    

    return insert_query
```

```python
cursor.execute(CreateTable(df_5, 'national_wealth_data'))
```
